```python
#https://plotly.com/python/
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a pandas DataFrame
data = {'year': [i for i in range(20)],
        'sales': np.random.randn(20)**2,
        'sal1': np.random.randn(20)**2,
        'sal2': np.random.randn(20)**2,
        'sal': np.random.randn(20)**2
        }
df = pd.DataFrame(data)

# Create a line plot
df.plot(x='year', y=[name for name in data.keys() if name != 'year'], kind='line')
plt.show()

# Create a bar chart
df.plot(x='year', y=[name for name in data.keys() if name != 'year'], kind='bar')
plt.show()

# Scatter
df.plot(x='year', y='sales', kind='scatter')
plt.show()

# Hist
df.plot(y=[name for name in data.keys() if name != 'year'], kind='hist')
plt.show()

```

```python
# boxplot
dic = {'A': np.random.randn(1000),
       'B': np.random.randn(1000)}
df = pd.DataFrame(dic)
# Create a boxplot
df.plot(kind='box')
plt.show()

df.plot(y = 'A',kind='hist')
plt.show()
```

```python
# Create a pandas DataFrame
data = {'year': [2015, 2016, 2017, 2018, 2019],
        'sales': [1000, 1200, 1300, 1400, 1600],
        'z' : [i for i in range(0,1000,200)]}
df = pd.DataFrame(data)

# Create an area plot
df.plot(x='year', y='sales', kind='area')
plt.show()
```

```python
import plotly.express as px

fig =px.scatter_3d(df, x='year',y='sales',z='z')
fig.show()
```
